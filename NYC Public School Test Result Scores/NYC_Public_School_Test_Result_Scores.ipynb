Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a maximum score of 800 points. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

We have been provided with a dataset called `schools.csv`, which is previewed below.

We have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

## What are the best math schools in NYC?

```python
best_math_schools = schools[schools["average_math"] >= 800*0.8]
best_math_schools = best_math_schools[["school_name", "average_math"]].sort_values("average_math", ascending=False)
best_math_schools
```

## What are the top ten SAT scoring schools in NYC?

```python
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
schools.sort_values("total_SAT", ascending=False, inplace=True)
top_10_schools = schools[["school_name", "total_SAT"]][:10]
top_10_schools
```

## Which borough of NYC has the largest standard deviation?

```python
largest_std_dev = schools.groupby("borough").agg(
    num_schools=("school_name", "count"), 
    average_SAT=("total_SAT", "mean"), 
    std_SAT=("total_SAT", "std")
).round(2)
largest_std_dev.sort_values("std_SAT", ascending=False, inplace=True)
largest_std_dev = largest_std_dev[:1]
largest_std_dev
```
